Q2. Most selling product?
a. by quantity
b. by revenue

```python
data.head()
```

```python
data.groupby('Product')
```

```python
data.groupby('Product')['Quantity'].sum()
```

```python
most_quantity = data.groupby('Product')['Quantity'].sum().sort_values(ascending=False)
```

```python
most_quantity
```

```python
most_quantity = most_quantity.reset_index()
most_quantity
```

```python
type(most_quantity)
```

```python
plt.figure(figsize=(9,4))
plt.bar(most_quantity['Product'], most_quantity['Quantity'], color=['red', 'green', 'blue'], width=0.4)
plt.title("Most Selling Product by Quantity")
plt.xlabel("Product")
plt.ylabel("Quantity")
plt.grid(False)
```

By Revenue

```python
data.head()
```

```python
data['Revenue'] = data['Price'] * data['Quantity']
```

```python
data.head()
```

```python
data.groupby('Product')['Revenue'].sum().sort_values(ascending=False)
```

```python
most_revenue = data.groupby('Product')['Revenue'].sum().sort_values(ascending=False)
```

```python
most_revenue
```

```python
most_revenue = most_revenue.reset_index()
```

```python
most_revenue
```

```python
plt.bar(most_revenue['Product'], most_revenue['Revenue'], color=['red', 'green', 'blue'], width=0.4)
plt.title("Most Selling Product by Revenue")
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.grid(False)
```
